The following measurements were recorded for
the drying time, in hours, of a certain brand of latex
paint.


```
3.4 2.5 4.8 2.9 3.6
2.8 3.3 5.6 3.7 2.8
4.4 4.0 5.2 3.0 4.8
```



Assume that the measurements are a simple random
sample.


1.   What is the sample size for the above sample?
2.   Calculate the sample mean for these data.
3.  Calculate the sample median.
4.  Plot the data by way of a dot plot.
5.  Compute the 20% trimmed mean for the above
data set.
6.  Is the sample mean for these data more or less de-
scriptive as a center of location than the trimmed
mean



# using Numpy libaray

````python
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [3.4, 2.5, 4.8, 2.9, 3.6, 2.8, 3.3, 5.6, 3.7, 2.8, 4.4, 4.0, 5.2, 3.0, 4.8]

# Sample size
sample_size = len(data)

# Sample mean
sample_mean = np.mean(data)

# Sample median
sample_median = np.median(data)

# Sorting the data
sorted_data = sorted(data)

# Creating a dot plot
plt.figure(figsize=(8, 4))
plt.plot(sorted_data, [1] * len(sorted_data), 'bo', markersize=8)  # 'bo' means blue color and circle marker
plt.yticks([])  # Hide y-axis ticks
plt.xlabel('Drying Time (hours)')
plt.title('Dot Plot of Drying Time')
plt.grid(True, axis='x')  # Show grid along x-axis
plt.show()

# 20% trimmed mean
trim_amount = int(0.2 * sample_size)  # Number of data points to trim
trimmed_data = sorted_data[trim_amount:-trim_amount]
trimmed_mean = np.mean(trimmed_data)

# Comparison of sample mean and trimmed mean
comparison = "Sample mean is more descriptive." if sample_mean > trimmed_mean else "Trimmed mean is more descriptive."

# Output
print("Sample Size:", sample_size)
print("Sample Mean:", sample_mean)
print("Sample Median:", sample_median)
print("20% Trimmed Mean:", trimmed_mean)
print("Comparison of sample mean and trimmed mean:", comparison)
````
